```python
def runif(): 
    """uniform random number generator"""
    return np.random.uniform(0, 1, 1)[0]
```

```python
def index_list(L): 
    """from a list, build a dictionary which list element to position"""
    return dict(zip(L, [x for x in range(0, len(L))]))
```

```python
Action = namedtuple('Action', ['weight'])
State = namedtuple('State', ['ranking'])
```

```python
def get_ranking(securities_returns, lookback, look_ahead = False):
    ranking = {"Date":[],"USD.EUR":[],"USD.CAD":[],"USD.JPY":[],"hour":[]}
    var_asset = {"Date":[],"USD.EUR":[],"USD.CAD":[],"USD.JPY":[], "hour":[]}
    dates = []
#     for i in range(23*60,len(securities_returns),23*60):
    # at what interval you want to train
    interval = 60
    for i in range(0,len(securities_returns),interval):
        vector = securities_returns[i:i+lookback]
#         print(vector.index[0],vector.index[-1])
#         prior_vector = securities_returns[i-interval:i+lookback-interval]
        var = vector.var().values
        rank = var.argsort().argsort()
#         rank = tuple(-1*rank + 2)
        if not look_ahead:
            ranking["Date"].append(vector.index[-1]) # 0: forward looking (cheating), -1: lookback
        else:
            ranking["Date"].append(vector.index[0])
        ranking["USD.EUR"].append(rank[0])
        ranking["USD.CAD"].append(rank[1])
        ranking["USD.JPY"].append(rank[2])
#         ranking["hour"].append(vector.index[-1].hour) # 0: forward looking (cheating), -1: lookback
        if not look_ahead:
            var_asset["Date"].append(vector.index[-1]) # 0: forward looking (cheating), -1: lookback
        else:
            var_asset["Date"].append(vector.index[0])
        var_asset["USD.EUR"].append(var[0])
        var_asset["USD.CAD"].append(var[1])
        var_asset["USD.JPY"].append(var[2])
#         var_asset["hour"].append(vector.index[-1].hour) # 0: forward looking (cheating), -1: lookback

        dates.append(vector.index[-1])
    ranking_df = pd.DataFrame(ranking, columns=['Date', 'USD.EUR', 'USD.CAD','USD.JPY'])
    ranking_df.set_index('Date', inplace=True)
    
    var_asset_df = pd.DataFrame(var_asset, columns=['Date', 'USD.EUR', 'USD.CAD','USD.JPY'])
    var_asset_df.set_index('Date', inplace=True)
    
    return ranking_df, var_asset_df
```

```python
class epoch_data(object):
    
    def __init__(self, benchmark_reward, rewards, test_result, time_elapsed, random_counter, index, train_result, train_result_oracle, test_result_oracle):
        self.benchmark_reward = np.array(benchmark_reward)
        self.rewards = np.array(rewards)
        self.test_result = test_result
        self.time_elapsed = time_elapsed
        self.random_counter = random_counter
        self.index = index
        self.train_result = train_result
        self.train_result_oracle = train_result_oracle
        self.test_result_oracle = test_result_oracle
```

```python
alpha = 0.01 ## learning rate

class TabularQLearner:
    def __init__(self, act_space, state_space, q_learning_function = "alpha",benchmark = "min_var",eps=0.1):
        self.actions = act_space
        self.states = state_space
        self.lookup = index_list(state_space)
        self.Q = np.zeros((len(self.states), len(self.actions)))
        self.epsilon = eps
        self.action_lookup = index_list(act_space)
        self.zero_trade = self.action_lookup[act_space[0]]
        self.q_learning_function = q_learning_function
        self.benchmark = benchmark
        self.epoch_data = []
        self.Q_matrixs = []
    def __greedy_action_index(self, state_index):
        """calculate the integer for indexing into self.Q or self.actions of the greedy action"""
        q = self.Q[state_index]
        if not np.any(q):
            return self.zero_trade
        else:
            return np.argmax(q)

    def greedy_action(self, state):
        """calculate the actual greedy action, not just its index"""
        return self.actions[self.__greedy_action_index(self.lookup[state])]

    def maxQ(self, state):
        return max(self.Q[self.lookup[state]])

    def train(self, rankings,epochs = 10):
        for e in range(epochs):
            random_counter = 0
            t0 = time.time()
            benchmark_reward = []
            rewards = []
            for r in range(0,len(rankings)-1):
                rank = tuple(rankings.loc[rankings.index[r]].values)
                curr = State(ranking=rank)

                i = self.lookup[curr]

                rand = runif()
                # with probability epsilon, select action randomly from amongst all available actions with equal probability
                if rand <= self.epsilon:
                    a = random.choice(self.actions)
                    random_counter += 1
                else:
                    a = self.greedy_action(curr)
                j = self.action_lookup[a]

                # for reward calculation
                train_df = ret_df_bps[(ret_df_bps.index <= rankings.index[r+1]) & (ret_df_bps.index >= rankings.index[r])]

                benchmark_var = self.benchmark_train_reward(train_df,benchmark = self.benchmark)
                benchmark_reward.append(benchmark_var)
            
                train_performance = np.array(a.weight)*train_df
                train_port = train_performance.sum(axis=1)
                reward = -train_port.var()
                rewards.append(reward)

                self.Q[i,j] = self.q_learning_update(self.q_learning_function, i, j, alpha, reward)
            
            time_for_epoch = time.time() - t0
            self.Q_matrixs.append(copy.deepcopy(self.Q))
#             current_epoch_data = self.get_epoch_data(e)
            test_result = self.get_test_result(self.benchmark)
            train_result = self.get_train_result(self.benchmark)
            
#             train_result_oracle = self.get_train_result_oracle()
#             test_result_oracle = self.get_test_result_oracle()
        
            curr_epoch_data = epoch_data(benchmark_reward, rewards, test_result, time_for_epoch, random_counter, rankings.index, train_result, None, None)
            self.epoch_data.append(curr_epoch_data)
        
        
    def q_learning_update(self,q_learning_function,i,j,alpha,reward):
        if q_learning_function == "alpha":
            return self.Q[i,j] + alpha * (reward - self.Q[i,j])
        if q_learning_function == "alpha-gamma":
            return self.Q[i,j] + alpha * (reward + 0.1*max(self.Q[i]) -self.Q[i,j])
        assert False, "Wrong Input Agent"
        
#     def get_epoch_data(self,epoch):
#         if epoch == 0:
#             return np.mean(self.Q_matrixs[epoch])
#         else:
#             currentQ = self.Q_matrixs[epoch]
#             prevQ = self.Q_matrixs[epoch - 1]
#             average_diff = np.mean(abs(currentQ - prevQ))
#             return average_diff
        
    def get_benchmark_weights(self,train_df,benchmark = "min_var"):
        if benchmark == "min_var":
            rank = np.array(rankings.loc[rankings.index == train_df.index[-1]].values)
            weight = np.array([np.min(weightOptions)]*len(weightOptions[0]))         
            weight[rank.argmin()] = np.max(weightOptions)         
            return weight
        if benchmark == "min_var_cheat":
#             print(train_df)
            rank = np.array(rankings_cheat.loc[rankings_cheat.index == train_df.index[0]].values)
            weight = np.array([np.min(weightOptions)]*len(weightOptions[0]))          
            weight[rank.argmin()] = np.max(weightOptions)
            return weight
        
    def benchmark_train_reward(self,train_df,benchmark = "min_var"):
        weight = self.get_benchmark_weights(train_df,benchmark)
        train_performance = np.array(weight)*train_df
        train_port = train_performance.sum(axis=1)
        reward = -train_port.var()
        return reward
    
    def get_ranking_data(self,train_test, benchmark):
        if train_test == "train" and benchmark == "min_var":
            return ranking_df_train
        if train_test == "train" and benchmark == "min_var_cheat":
            return ranking_df_train_cheat
        if train_test == "test" and benchmark == "min_var":
            return ranking_df_test
        if train_test == "test" and benchmark == "min_var_cheat":
            return ranking_df_test_cheat
        

    def get_test_result(self, benchmark):
        test_return = pd.DataFrame()
        interval = 1
        ranking_df = self.get_ranking_data("test", benchmark)
        for i in range(0,len(ranking_df)-1,interval):
            rank = tuple(ranking_df.loc[ranking_df.index[i]].values)
        #     print(rank,hour)

            curr = State(ranking =rank)
            optimal_weight = self.greedy_action(curr).weight

            test_df = ret_df[(ret_df_bps.index <= ranking_df.index[i+1]) & (ret_df_bps.index >= ranking_df.index[i])]

            test_performance = list(optimal_weight)*test_df
            test_port = test_performance.sum(axis=1)
            test_return = test_return.append(pd.DataFrame(test_port))
        return test_return.std().values[0]*np.sqrt(23*60*252)
    def get_train_result(self, benchmark):
        test_return = pd.DataFrame()
        interval = 1
        ranking_df = self.get_ranking_data("train", benchmark)
        for i in range(0,len(ranking_df)-interval,interval):
            rank = tuple(ranking_df.loc[ranking_df.index[i]].values)

            curr = State(ranking =rank)
            optimal_weight = self.greedy_action(curr).weight
#             self.actions

            train_df = ret_df[(ret_df_bps.index <= ranking_df.index[i+interval]) & (ret_df_bps.index >= ranking_df.index[i])]

            test_performance = list(optimal_weight)*train_df
            test_port = test_performance.sum(axis=1)
            test_return = test_return.append(pd.DataFrame(test_port))
        return test_return.std().values[0]*np.sqrt(23*60*252)
    
    def get_train_result_oracle(self):
        test_return = pd.DataFrame()
        interval = 1
        ranking_df = self.get_ranking_data("train", "min_var_cheat")
        for i in range(0,len(ranking_df)-interval,interval):
            rank = tuple(ranking_df.loc[ranking_df.index[i]].values)

            curr = State(ranking =rank)
            optimal_weight = self.greedy_action(curr).weight
#             self.actions

            train_df = ret_df[(ret_df_bps.index <= ranking_df.index[i+interval]) & (ret_df_bps.index >= ranking_df.index[i])]

            test_performance = list(optimal_weight)*train_df
            test_port = test_performance.sum(axis=1)
            test_return = test_return.append(pd.DataFrame(test_port))
        return test_return.std().values[0]*np.sqrt(23*60*252)
    
    def get_test_result_oracle(self):
        test_return = pd.DataFrame()
        interval = 1
        ranking_df = self.get_ranking_data("test", "min_var_cheat")
        for i in range(0,len(ranking_df)-interval,interval):
            rank = tuple(ranking_df.loc[ranking_df.index[i]].values)

            curr = State(ranking =rank)
            optimal_weight = self.greedy_action(curr).weight
#             self.actions

            train_df = ret_df[(ret_df_bps.index <= ranking_df.index[i+interval]) & (ret_df_bps.index >= ranking_df.index[i])]

            test_performance = list(optimal_weight)*train_df
            test_port = test_performance.sum(axis=1)
            test_return = test_return.append(pd.DataFrame(test_port))
        return test_return.std().values[0]*np.sqrt(23*60*252)
            
        
        
```

```python
def get_ranking_data(train_test, benchmark):
    if train_test == "train" and benchmark == "min_var":
        return ranking_df_train
    if train_test == "train" and benchmark == "min_var_cheat":
        return ranking_df_train_cheat
    if train_test == "test" and benchmark == "min_var":
        return ranking_df_test
    if train_test == "test" and benchmark == "min_var_cheat":
        return ranking_df_test_cheat

def benchmark_train_reward(train_df,benchmark):
    weight = self.get_benchmark_weights(train_df,benchmark)
    train_performance = np.array(weight)*train_df
    train_port = train_performance.sum(axis=1)
    reward = -train_port.var()
    return reward
def absolute_oracle(train_test):
    test_return = pd.DataFrame()
    interval = 1
    ranking_df = get_ranking_data(train_test, "min_var_cheat")
    for i in range(0,len(ranking_df)-interval,interval):
#         print(ranking_df.index[i])
        rank = np.array(tuple(ranking_df.loc[ranking_df.index[i]].values))
#         print(rank)
        weight = np.array([np.min(weightOptions)]*len(weightOptions[0]))          
        weight[rank.argmin()] = np.max(weightOptions)

        train_df = ret_df[(ret_df_bps.index <= ranking_df.index[i+interval]) & (ret_df_bps.index >= ranking_df.index[i])]

        test_performance = list(weight)*train_df
        test_port = test_performance.sum(axis=1)
        test_return = test_return.append(pd.DataFrame(test_port))
    return round(test_return.std().values[0]*np.sqrt(23*60*252),6)

def markowitz_weight(lookback_df):
    mu = pyp.expected_returns.mean_historical_return(lookback_df, returns_data=True)
    S = np.cov(lookback_df.T, ddof=0)
    ef = pyp.efficient_frontier.EfficientFrontier(mu, S, weight_bounds=(0,1))
    weights = ef.min_volatility()
    return weights.values()
def absolute_markowitz(train_test, use_action_space = False):
    test_return = pd.DataFrame()
    interval = 1
    ranking_df = get_ranking_data(train_test, "min_var")
    for i in range(0,len(ranking_df)-interval,interval):
        train_df = ret_df[(ret_df_bps.index <= ranking_df.index[i+interval]) & (ret_df_bps.index >= ranking_df.index[i])]

        weight = markowitz_weight(train_df)
        if use_action_space:
            action_space_weight = np.array([np.min(weightOptions)]*len(weightOptions[0]))
            action_space_weight[np.array(list(weight)).argmax()] = np.max(weightOptions)
            weight = action_space_weight

        test_performance = list(weight)*train_df
        test_port = test_performance.sum(axis=1)
        test_return = test_return.append(pd.DataFrame(test_port))
    return round(test_return.std().values[0]*np.sqrt(23*60*252),6)
```
